### OpenCV

```python
## PILLOW: RGB
## OpenCV: BGR
```

```python
import cv2
```

```python
img=cv2.imread("dog.jpg")
plt.imshow(img)
```

```python
type(img)
```

```python
new_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

```python
gray_img=cv2.cvtColor(new_img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img)
plt.show()
```

```python
img_flip=cv2.flip(new_img,0)
plt.imshow(img_flip)
```

```python
img_flip=cv2.rotate(new_img,cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_flip)
```

```python
img.shape
```

```python
img.max()
```

```python
img.min()
```

```python
blue,green,red=new_img[:,:,0],img[:,:,1],img[:,:,2]
```

```python
plt.imshow(blue)
```

```python
crop_image=new_img[20:140,50:250,:]
plt.imshow(crop_image)
```

```python
top=40
bottom=80
left=75
right=150
eye_img=crop_image[top:bottom,left:right,:]
plt.imshow(eye_img)
```

```python
### changing image pixel values
pixel_img=(new_img[0:,250:]=255)
plt.imshow(pixel_img)
```

```python
eye_img[0:,60:]=255
plt.imshow(eye_img)    ### 255=white
```

```python
eye_img[0:,60:,0]=255     ### accessing the red channel
plt.imshow(eye_img)    ### 255=white
```

```python
import numpy as np
upper=60
lower=120
left=100
right=200
start_point,end_point=(left,upper),(right,lower)
image_draw=np.copy(new_img)
cv2.rectangle(image_draw,pt1=start_point,pt2=end_point,color=(0,255,0),thickness=3)
plt.imshow(image_draw)
```
